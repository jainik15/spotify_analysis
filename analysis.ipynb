```python
import pandas as pd

df= pd.read_csv("spotify_2023.csv")
print("-First 5 rows-")
display(df.head())

print("-Data types and info-")
df.info()
```

```python
df['in_shazam_charts'] = df['in_shazam_charts'].fillna(0)
df['in_shazam_charts'] = df['in_shazam_charts'].str.replace(',','')
df['in_shazam_charts'] = pd.to_numeric(df['in_shazam_charts'])

df['in_deezer_charts'] = df['in_deezer_charts'].fillna(0)
df['in_deezer_charts'] = df['in_deezer_charts'].str.replace(',','')
df['in_deezer_charts'] = pd.to_numeric(df['in_deezer_charts'])

print('---Updated data types---')
df.info()

print("---First Five Rows---")
display(df.head())
```

```python
import matplotlib.pyplot as plt

artist_count = df['artist(s)_name'].value_counts()
print('--Top Artitst By Number Of SOng--')
print(artist_count)

plt.figure(figsize=(10,6))

artist_count.plot(kind='bar', color='skyblue')
plt.title("Number of top songs by artist in 2023")
plt.xlabel('Artist(s)')
plt.ylabel('Number of songs in top list')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

print("--Statistics of Dancebility--")
print(df['danceability_%'].describe())

plt.figure(figsize=(10,6))

plt.hist(df['danceability_%'], bins=10, color='mediumpurple', edgecolor='black')

plt.title('Distribution of Danceability Scores in Top Songs')
plt.xlabel('Danceability %')
plt.ylabel('Number of Songs (frequency)')

plt.axvline(df["danceability_%"].mean(), color = 'red', linestyle='dashed', linewidth=2, label=f"Mean: {df['danceability_%'].mean(): .2f}")
plt.axvline(df["danceability_%"].median(), color = 'blue', linestyle='dashed', linewidth=2, label=f"Median: {df['danceability_%'].median(): .2f}")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(df['energy_%'], df['valence_%'], color='teal', alpha=0.7)
plt.title('Energy vs. Valence of Top Spotify Songs')
plt.xlabel('Energy %')
plt.ylabel('Valent (Positiveness) %')

plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt

key_mode_count = df['mode'].value_counts()

print("---Song Count By Key Note---")
print(key_mode_count)

plt.figure(figsize=(8,8))
plt.pie(key_mode_count, labels=key_mode_count.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue'])

plt.title('Proportion of songs in Major vs Minor keys \n')
plt.axis('equal')

plt.show()
```
